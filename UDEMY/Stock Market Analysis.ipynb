```python
days = 365
dt = 1/days
mu = rets.mean()['GOOG']
sigma = rets.std()['GOOG']
```

```python
def stock_monte_carlo(start_price,days,mu,sigma):
    price = np.zeros(days)
    price[0] = start_price
    shock = np.zeros(days)
    drift = np.zeros(days)
    
    for x in xrange(1,days):
        shock[x] = np.random.normal(loc=mu*dt,scale=sigma*np.sqrt(dt))
        drift[x] = mu*dt
        price[x] = price[x-1] + (price[x-1]*(drift[x]+shock[x]))
    return price
```

```python
GOOG.head()
```

```python
start_price = 1212.989

for run in xrange(100):
    plt.plot(stock_monte_carlo(start_price,days,mu,sigma))
plt.xlabel('Days')
plt.ylabel('Price')
plt.title('Monte Carlo Analysis for Google')
```

```python
runs = 100000

simulation = np.zeros(runs)

for run in xrange(runs):
    simulation[run] = stock_monte_carlo(start_price,days,mu,sigma)[days-1]
```

```python
q = np.percentile(simulation,1)

plt.hist(simulation,bins=2000,color='blue')


# Starting Price
plt.figtext(0.6, 0.8, s="Start price: $%.2f" %start_price,color='blue')
# Mean ending price
plt.figtext(0.6, 0.7, "Mean final price: $%.2f" % simulation.mean(),color='blue')

# Variance of the price (within 99% confidence interval)
plt.figtext(0.6, 0.6, "VaR(0.99): $%.2f" % (start_price - q,),color='blue')

# Display 1% quantile
plt.figtext(0.15, 0.6, "q(0.99): $%.2f" % q,color='blue')

# Plot a line at the 1% quantile result
plt.axvline(x=q, linewidth=4, color='r')

# Title
plt.title(u"Final price distribution for Google Stock after %s days" % days, weight='bold');
```
